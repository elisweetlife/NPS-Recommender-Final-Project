```python
# Dependencies
import requests
from pprint import pprint
import pandas as pd

url = "https://developer.nps.gov/api/v1/alerts?limit=1000&api_key=TyuZrBgyshbHS2txyIXVovjDPtLAp8jYUJPSa1hf"
```

```python
# Request alerts
response = requests.get(url).json()
```

```python
response
```

```python
alert_list = response["data"]
pprint(alert_list)
```

```python
category = []
description = []
id = []
parkCode = []
title = []
url = []
```

```python
for alert in alert_list:
    category.append(alert["category"])
    description.append(alert["description"])
    id.append(alert["id"])
    parkCode.append(alert["parkCode"])
    title.append(alert["title"])
    url.append(alert["url"])
```

```python
alert_dict = {
    "category": category,
    "description": description,
    "id": id,
    "parkCode": parkCode,
    "title": title,
    "url": url
}
alert_data = pd.DataFrame(alert_dict)
alert_data.head()
```

```python
alert_data.to_csv("alerts.csv")
```
