```python
import pandas as pd
```

```python
data1=pd.read_csv("39.562141,-77.853069.csv")
```

```python
data1
```

```python
extracted_data=data1[["date_time", "maxtempC", "mintempC", "humidity", "visibility", "windspeedKmph", "DewPointC", "cloudcover"]]
```

```python
extracted_data.head()
```

```python
extracted_data['Year'] = extracted_data['date_time'].str[:4]
extracted_data['Month'] = extracted_data['date_time'].str[5:7]
```

```python
extracted_data.head(20)
```

```python
extracted_data['Qtr'] = pd.to_datetime(extracted_data.date_time).dt.quarter
```

```python
extracted_data.head()
```

```python
final_data=extracted_data.groupby(['Year', 'Qtr']).mean() 
```

```python
final_data
```

```python
final_data=final_data.reset_index()
```
