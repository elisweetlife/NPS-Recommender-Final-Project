```python
# Initialize empty lists for storing JSON responses
totalSites = []
description = []
id = []
latitude = []
longitude = []
name = []
numberOfSitesFirstComeFirstServe = []
numberOfSitesReservable = []
operatingHours = []
parkCode = []
reservationInfo = []
reservationUrl = []
```

```python
for list in parkCodeLists:
    url = "https://developer.nps.gov/api/v1/campgrounds?parkCode=" + list + "&api_key=" + apiKey
    response = requests.get(url).json()
    campground_list = response["data"]
    print(list)
    print(len(campground_list))
          
    for campground in campground_list:
        totalSites.append(campground["campsites"]["totalSites"])
        description.append(campground["description"])
        id.append(campground["id"])
        latitude.append(campground["latitude"])
        longitude.append(campground["longitude"])
        name.append(campground["name"])  
        numberOfSitesFirstComeFirstServe.append(campground["numberOfSitesFirstComeFirstServe"])
        numberOfSitesReservable.append(campground["numberOfSitesReservable"])
        operatingHours.append(campground["operatingHours"])
        parkCode.append(campground["parkCode"])
        reservationInfo.append(campground["reservationInfo"])
        reservationUrl.append(campground["reservationUrl"])
```

```python
campground_dict = {
    "parkCode": parkCode,
    "name": name,
    "totalSites": totalSites,
    "latitude": latitude,
    "longitude": longitude,
    "numberOfSitesFirstComeFirstServe": numberOfSitesFirstComeFirstServe,
    "numberOfSitesReservable": numberOfSitesReservable,
    "description": description,
    "operatingHours": operatingHours,
    "reservationInfo": reservationInfo,
    "id": id,
    "reservationUrl": reservationUrl 
}
campground_data = pd.DataFrame(campground_dict)
campground_data.head()
```
